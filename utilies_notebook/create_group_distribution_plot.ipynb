```python
import csv
```

```python
group_sizes = []
with open('group_size_distribution.csv', newline='', encoding='utf-8-sig', errors='ignore') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';')
    for index, row in enumerate(spamreader):
            group_sizes.append(row)
```

```python
import matplotlib.pyplot as plt
import numpy as np

for index, element in enumerate(group_sizes):
    group_sizes[index] = [int(element[0]), int(element[1])]

number_of_samples = [i[1] for i in group_sizes]
size_of_samples = [i[0] for i in group_sizes]

y_pos = np.arange(len(group_sizes))

plt.figure(figsize=(30,10))
# Create bars
plt.bar(y_pos, number_of_samples, label = "Number of samples")
plt.xticks(fontsize=10)
plt.yticks(fontsize=20)

# Create names on the x-axis
plt.xticks(y_pos, size_of_samples)

plt.xlabel('Group Size', fontsize=20)
plt.ylabel('Number of samples', fontsize=20)


plt.legend(loc="upper right", fontsize=30)

 
# Show graphic
plt.show()
```
