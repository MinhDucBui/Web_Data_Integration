```python
import csv
```

```python
book_crossing = []
with open('datasets/BX-CSV-Dump/BX-Books.csv', newline='', encoding='utf-8', errors='ignore') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';')
    for index, row in enumerate(spamreader):
        if index > 0:
            book_crossing.append(row[0])
```

```python
goodread = []
with open('datasets/goodread.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    for index, row in enumerate(spamreader):
        if index > 0:
            goodread.append(row[4])
```

```python
goodread_31 = []
with open('datasets/goodreads_books.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    for index, row in enumerate(spamreader):
        if index > 0:
            goodread_31.append(row[20])
```

```python
goodread_bestbook_train = []
with open('goodread_bestbook_training_set.csv', newline='', encoding='utf-8', errors='ignore') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';')
    for index, row in enumerate(spamreader):
            goodread_bestbook_train.append(row[0])
```

```python
def intersection(lst1, lst2): 
    lst3 = [value for value in lst1 if value in lst2] 
    return lst3 
```
